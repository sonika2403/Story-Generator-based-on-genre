**CREATIVE STORY GENERATOR BASED ON GENRE**

The core concept of this project is to build a story generator, which will take user inputs such as choosing a genre, giving the initial prompt based on which they want the story. This generator gives a story based on the requirements. The project used GPT-NEO model and hugging face transformers library, python code for implementation. Try it out!

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
#Using the model GPT NEO
#Loading the model and the tokenizer.
model_name = "EleutherAI/gpt-neo-1.3B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def get_genre_prompt(genre):
  #Defining genre keywords so that the assisstant can identify
    genre_prompts = {
        "fantasy": "In a magical kingdom, full of dragons, wizards, and mythical creatures, ",
        "sci-fi": "In a futuristic world, where technology reigns and space travel is common, ",
        "mystery": "In a dark, eerie town, where every corner hides a secret, ",
        "romance": "In a quaint, beautiful city, where love blossoms under the stars, ",
        "adventure": "In a dense jungle, filled with ancient ruins and hidden treasures, "
    }

    # This code tried to return appropriate story if the prompt completely does not match
    return genre_prompts.get(genre.lower(), "Once upon a time, ")
```

```python
def generate_story_with_genre(prompt, max_length=150):
    inputs = tokenizer(prompt, return_tensors="pt")

    #main constraints to look into while generating a story
    outputs = model.generate(
        inputs["input_ids"],
        max_length=max_length,
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        top_p=0.95,
        temperature=0.7
    )

    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_text
```

```python
#The main function which will take the user input tp generate a proper story
def main():
    print("Welcome to the Story Generator!")
    print("Choose a genre for your story:")
    print("1. Fantasy")
    print("2. Sci-Fi")
    print("3. Mystery")
    print("4. Romance")
    print("5. Adventure")

    #This code will ask to choose a number corresponding to genre to choose a genre of user choice.
    genre_choice = input("Enter the number corresponding to your genre choice: ")

    #This will be the genre listed below
    genres = ["fantasy", "sci-fi", "mystery", "romance", "adventure"]
    selected_genre = genres[int(genre_choice) - 1]

    user_prompt = input(f"Enter the beginning of your story for the {selected_genre} genre: ")
    genre_prompt = get_genre_prompt(selected_genre)
    full_prompt = genre_prompt + user_prompt
    print("\nGenerating story continuation...\n")

    #This below code will generate and display the continued story.
    generated_story = generate_story_with_genre(full_prompt, max_length=150)
    print("\nStory generated:\n")
    print(generated_story)

#This runs the main function
if __name__ == "__main__":
    main()
```
